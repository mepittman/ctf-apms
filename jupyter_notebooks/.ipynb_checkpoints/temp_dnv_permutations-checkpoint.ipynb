```python
def load_interactome(int_type):
    gene_file = "{}/ntermediate/interactome_lists/{}_genes.txt".format(base_path,int_type)
    gene_list = []
    
    with open(gene_file) as f:
        for line in f:
            gene_list.append(line.strip())
            
    return gene_list
```

```python
# Random shuffle using numpy
def shuffle_dict_keys(a):
    keys = list(a.keys())
    np.random.shuffle(keys)
    b = dict(zip(keys, a.values()))
    return(b)


# Scramble the IDs from the case and control dictionaries
def scramble_ids(case_dict, ctrl_dict):
    
    combined_dict = {**case_dict, **ctrl_dict} 
    scrambled_dict = shuffle_dict_keys(combined_dict)
    
    case_ids = []
    ctrl_ids = []
    for key in combined_dict:
        if key.startswith("1-") or key.startswith("GT"):
            case_ids.append(key)
        else:
            ctrl_ids.append(key)
    
    scram_case = {k: scrambled_dict[k] for k in case_ids}
    scram_ctrl = {k: scrambled_dict[k] for k in ctrl_ids}
    
    return(scram_case, scram_ctrl)

    
# Function to return n permuted odds ratios
def permute_case_ctrl(case_dict, ctrl_dict, gene_list, n_perm):
    
    perm_list = []
    contingency_list = []
    
    for i in range(0,n_perm):
        scram_case, scram_ctrl = scramble_ids(case_dict, ctrl_dict)
        contingency = get_OR(scram_case, scram_ctrl, gene_list)
        contingency_list.append(contingency)
        perm_list.append(contingency[5])
        
    contingency_df = pd.DataFrame(contingency_list, 
                                  columns=['case_intact_count', 'case_nonintact_count', 
                                           'ctrl_intact_count', 'ctrl_nonintact_count', 
                                           'OR', 'synonymous-corrected OR'])
        
    return(perm_list, contingency_df)
    
    
# Function to calculate p-value from list of permutation results
def perm_result(perm_list, true_OR):
    
    nGreater = len([num for num in perm_list if num > true_OR])
    pval = nGreater/len(perm_list)
    minp = 1/len(perm_list)
    
    if nGreater == 0:
        pval = "<{}".format(minp)
        
    return(pval)
```

```python
def get_id_dict(filename, mut_type, unknown_only = False):
    
    mut_str_dict = {'LoF_':['non','frameshift','startloss','stoploss'],
                    'splice_':['splice'], 'syn_':['syn'],
                   'allmuts_':['non','frameshift','startloss','stoploss','misD','splice','mis'],
                   'allmis_':['misD','mis']}
    
    id_dict = defaultdict(list)
    with open(filename) as f:
        
        first_line = f.readline().split(',')
        gene_idx = next(x for x,val in enumerate(first_line) if val == "Gene")
        class_idx = next(x for x,val in enumerate(first_line) if val in ['Variant.Class', 'Variant_Class'])
        next(f)
        
        for line in f:
            line_array = line.split(',')
            
            # Ignore if it's a known gene and we're ignoring them
            if unknown_only==True:
                if line_array[gene_idx] in known_genes:
                    continue
            # Only save to dictionary if it's in the type of variants we're analyzing
            if line_array[class_idx] in mut_str_dict[mut_type]:
                id_dict[line_array[0]].append(line_array[gene_idx])
                
    return id_dict
            
def add_placeholders(id_dict, expected_n):
    
    # Add empty placeholders for individuals with no called de novos in either category
    if len(id_dict) < expected_n:
        n_placeholders = expected_n - len(id_dict)
        ids = range(0,n_placeholders)
        for i in ids:
            id_dict["placeholder{}".format(i)] = ([],[])
            
    print(len(id_dict))
    return(id_dict)
    
def combine_id_dicts(nonsyn, syn, expected_n):
    id_list = [nonsyn.keys(), syn.keys()]
    ids_with_data = set([item for sublist in id_list for item in sublist])
    new_dict = {}
    
    for pid in ids_with_data:
        if pid not in nonsyn:
            nonsyn[pid] = []
        elif pid not in syn:
            syn[pid] = []
        new_dict[pid] = (nonsyn[pid], syn[pid])
    
    id_dict = add_placeholders(new_dict, expected_n)
    return id_dict
    
def get_dnv_dicts(mut_type, unknown_only = False):

    nonsyn_case_file = "{}/input/variants/nonsynonymous_dnvs_cases.csv".format(base_path)
    nonsyn_ctrl_file = "{}/input/variants/nonsynonymous_dnvs_ctrls.csv".format(base_path)
    syn_case_file = "{}/input/variants/synonymous_dnvs_cases.csv".format(base_path)
    syn_ctrl_file = "{}/input/variants/synonymous_dnvs_ctrls.csv".format(base_path)
        
    CHD_nonsyn_dict = get_id_dict(nonsyn_case_file, mut_type, unknown_only = unknown_only)
    ASD_nonsyn_dict = get_id_dict(nonsyn_ctrl_file, mut_type, unknown_only = unknown_only)
    CHD_syn_dict = get_id_dict(syn_case_file, "syn_", unknown_only = unknown_only)
    ASD_syn_dict = get_id_dict(syn_ctrl_file, "syn_", unknown_only = unknown_only)

    CHD_proband_dict = combine_id_dicts(CHD_nonsyn_dict, CHD_syn_dict, n_case_trios)
    ASD_sibling_dict = combine_id_dicts(ASD_nonsyn_dict, ASD_syn_dict, n_ctrl_trios)
 
    return CHD_proband_dict, ASD_sibling_dict
```

```python
# Functions to calculate odds ratio
def get_OR(case_dict, ctrl_dict, gene_list, truth = False):
    
    case_ct, case_non_ct = get_ctable(case_dict, gene_list)
    ctrl_ct, ctrl_non_ct = get_ctable(ctrl_dict, gene_list)
                
    # Add pseudocount if any values are equal to zero
    if 0 in [case_ct, case_non_ct, ctrl_ct, ctrl_non_ct]:
        case_ct+=0.5; case_non_ct+=0.5; ctrl_ct+=0.5; ctrl_non_ct+=0.5;
        
    # Calculate odds ratio
    odds = (case_ct * ctrl_non_ct) / (ctrl_ct * case_non_ct)
    
    # Calculate the synonymous correction, and print it out
    syn_case_ct, syn_case_non_ct = get_ctable(case_dict, gene_list, var_type="synonymous")
    syn_ctrl_ct, syn_ctrl_non_ct = get_ctable(ctrl_dict, gene_list, var_type="synonymous")
    # Add pseudocount if any values are equal to zero
    if 0 in [syn_case_ct, syn_case_non_ct, syn_ctrl_ct, syn_ctrl_non_ct]:
        syn_case_ct+=0.5; syn_case_non_ct+=0.5; syn_ctrl_ct+=0.5; syn_ctrl_non_ct+=0.5;
    syn_odds = (syn_case_ct * syn_ctrl_non_ct) / (syn_ctrl_ct * syn_case_non_ct)
    corrected_odds = odds * 1/syn_odds
        
    # Print contingency table, to look at true OR numbers
    if truth == True:
        print("Case interactome muts: {}".format(case_ct))
        print("Case non-interactome muts: {}".format(case_non_ct))
        print("Ctrl interactome muts: {}".format(ctrl_ct))
        print("Ctrl non-interactome muts: {}".format(ctrl_non_ct))
        print("Odds ratio: {}".format(odds))
        
        print("Synonymous OR: ", syn_odds)
        print("Corrected OR: ", corrected_odds)
        
    return (case_ct, case_non_ct, ctrl_ct, ctrl_non_ct, odds, corrected_odds)

# Function to iterate through dictionary, returning contingency table
def get_ctable(dictionary, gene_list, var_type="non-synonymous"):
    yes_ct = 0; no_ct = 0
    
    if var_type=="non-synonymous":
        idx=0
    else:
        idx=1
    
    # Create contingency table - get yes and no values
    for indi in dictionary:
        
        for genes in dictionary[indi][idx]:

            if (genes in gene_list):
                yes_ct += 1
            else:
                no_ct += 1
                
    return(yes_ct, no_ct)
```

```python
def get_summary_stats(contingency_df):
        
    perm_list_original = list(contingency_df['OR'])
    perm_list_corrected = list(contingency_df['synonymous-corrected OR'])
    
    OR_original = perm_list_original.pop()
    OR_corrected = perm_list_corrected.pop()
    
    pval_original = perm_result(perm_list_original, OR_original)
    pval_corrected = perm_result(perm_list_corrected, OR_corrected)
    
    return [(perm_list_original, OR_original, pval_original), 
            (perm_list_corrected, OR_corrected, pval_corrected)]
```

```python
def visualize_perm(perm_list, OR, pval, plot_title=None, fname=None):

    # Show plot
    plt.clf()
    plt.hist(perm_list)
    plt.suptitle('{}'.format(plot_title), fontsize=18)
    plt.figtext(0.99, 0.01, 'p-value = {}'.format(pval), horizontalalignment='right')
    plt.xlabel('Odds Ratios', fontsize=16)
    plt.ylabel('Frequency', fontsize=16)
    ax = plt.axvline(OR, color='red', linewidth=2)
    plt.show()
    #plt.savefig(fname))
```

```python
def permutation_routine(int_type):
    
    print("Interactome: ", int_type)
    interactome_dict = {}

    interactome = load_interactome(int_type)
    print(len(interactome))
    
    # Print pvalues, save perm_lists for different stratifications of LoF minor allele frequency (MAF)
    for mut_type in ['LoF_', 'splice_', 'allmuts_', 'allmis_']:
        
        print(mut_type)
        
        for unknown in [True, False]:
            
            if unknown == True:
                known_str = 'unknownGenes'
            else:
                known_str = 'allGenes'

            # Reduce dictionaries to desired MAF cutoff
            probands, siblings = get_dnv_dicts(mut_type, unknown_only = unknown)

            # Get true odds ratio
            contingency = get_OR(probands, siblings, interactome, truth=True)
            true_OR = contingency[4]
            fixed_OR = contingency[5]

            # Get 1000 permutations and read out results
            perm_list, contingency_df = permute_case_ctrl(probands,siblings,interactome,n_perm)
            contingency_df = contingency_df.rename('Perm{}'.format)
            contingency_df.loc[len(contingency_df)] = contingency
            contingency_df.rename({contingency_df.index[-1]: "True"}, inplace=True)

            # Save out file
            out_path = '{}/{}'.format(out_base, int_type)
            if not os.path.exists(out_path):
                os.makedirs(out_path)
            contingency_df.to_csv('{}/{}_{}{}_permutation_contingency_data.csv'.format(
                out_path, int_type, mut_type, known_str))

            # Get summary stats
            summary_stats = get_summary_stats(contingency_df)
            original_list_or_pval = summary_stats[0]
            corrected_list_or_pval = summary_stats[1]

            dict_str = '{}_{}{}_original'.format(int_type, mut_type, known_str)
            interactome_dict[dict_str] = {"OR":summary_stats[0][1], "PVAL":summary_stats[0][2]}
            dict_str = '{}_{}{}_syn-corrected'.format(int_type, mut_type, known_str)
            interactome_dict[dict_str] = {"OR":summary_stats[1][1], "PVAL":summary_stats[1][2]}

            # Visualize
            #visualize_perm(original_list_or_pval[0],
            #               original_list_or_pval[1],
            #               original_list_or_pval[2],
            #               plot_title="{} {} mutation permutation - Uncorrected".format(int_type, mut_type))

            #visualize_perm(corrected_list_or_pval[0],
            #               corrected_list_or_pval[1],
            #               corrected_list_or_pval[2],
            #               plot_title="{} {} mutation permutation - Corrected".format(int_type, mut_type))
        
    return(interactome_dict)
```

```python
# Dot plot visualization
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt


class MidpointNormalize(mpl.colors.Normalize):
    def __init__(self, vmin, vmax, midpoint=0.05, clip=False):
        self.midpoint = midpoint
        mpl.colors.Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        normalized_min = max(0, 1 / 2 * (1 - abs((self.midpoint - self.vmin) / (self.midpoint - self.vmax))))
        normalized_max = min(1, 1 / 2 * (1 + abs((self.vmax - self.midpoint) / (self.midpoint - self.vmin))))
        normalized_mid = 0.5
        x, y = [self.vmin, self.midpoint, self.vmax], [normalized_min, normalized_mid, normalized_max]
        return sp.ma.masked_array(sp.interp(value, x, y))


vals = sp.array([[-5., 0], [5, 10]]) 
vmin = vals.min()
vmax = vals.max()

norm = MidpointNormalize(vmin=vmin, vmax=vmax, midpoint=0)
cmap = 'RdBu_r' 
```

```python
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt


class MidpointNormalize(mpl.colors.Normalize):
    def __init__(self, vmin, vmax, midpoint=0.05, clip=False):
        self.midpoint = midpoint
        mpl.colors.Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        normalized_min = max(0, 1 / 2 * (1 - abs((self.midpoint - self.vmin) / (self.midpoint - self.vmax))))
        normalized_max = min(1, 1 / 2 * (1 + abs((self.vmax - self.midpoint) / (self.midpoint - self.vmin))))
        normalized_mid = 0.5
        x, y = [self.vmin, self.midpoint, self.vmax], [normalized_min, normalized_mid, normalized_max]
        return sp.ma.masked_array(sp.interp(value, x, y))


vals = sp.array([[-5., 0], [5, 10]]) 
vmin = vals.min()
vmax = vals.max()

norm = MidpointNormalize(vmin=vmin, vmax=vmax, midpoint=0)
cmap = 'RdBu_r' 
```

```python
def get_unknown_dicts(mut_type):
    
    mut_str_dict = {'LoF_':['non','frameshift','startloss','stoploss'],
                    'misD_':['misD'], 'splice_':['splice'], 'mis_':['mis'],
                   'allmuts_':['non','frameshift','startloss','stoploss','misD','splice','mis'],
                   'allmis_':['misD','mis']}
    
    CHD_proband_dict = defaultdict(list)
    ASD_sibling_dict = defaultdict(list)
    
    if mut_type == "syn_":
        case_file = "/pollard/home/mpittman/APMS2/input/variants/NatGen/syn_DNV_cases.csv"
        ctrl_file = "/pollard/home/mpittman/APMS2/input/variants/NatGen/syn_DNV_ctrls.csv"
        
        with open(case_file) as f:
            next(f)
            for line in f:
                line_array = line.split(',')
                
                if line_array[9] in known_genes:
                    continue

                CHD_proband_dict[line_array[0]].append(line_array[9])

        with open(ctrl_file) as f:
            next(f)
            for line in f:
                line_array = line.split(',')
                
                if line_array[6] in known_genes:
                    continue

                ASD_sibling_dict[line_array[0]].append(line_array[6])
                    
    else:
        # Read through all LoF files, creating a dictionary of {ID:[variant_keys list], ID:[variant_keys list]}
        case_file = "/pollard/home/mpittman/APMS2/input/variants/NatGen_unpolished/DNV_cases.csv"
        ctrl_file = "/pollard/home/mpittman/APMS2/input/variants/NatGen_unpolished/DNV_ctrls.csv"

        with open(case_file) as f:
            next(f)
            for line in f:
                line_array = line.split(',')
                
                if line_array[9] in known_genes:
                    continue

                if line_array[10] in mut_str_dict[mut_type]:
                    CHD_proband_dict[line_array[0]].append(line_array[9])

        with open(ctrl_file) as f:
            next(f)
            for line in f:
                line_array = line.split(',')
                
                if line_array[6] in known_genes:
                    continue

                if line_array[7] in mut_str_dict[mut_type]:
                    ASD_sibling_dict[line_array[0]].append(line_array[6])

    return CHD_proband_dict, ASD_sibling_dict
```
