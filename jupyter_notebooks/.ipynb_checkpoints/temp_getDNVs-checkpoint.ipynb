```python
dat_dir = "/pollard/data/genetics/PCGC/WES/Yale/rare_PCGC_072020/pkl"

dnv_list= []

import os
import pickle as pkl
for file in os.listdir(dat_dir):
    with open('{}/{}'.format(dat_dir,file),'rb') as p:
        GT_dict = pkl.load(p)
        for key, value in GT_dict.items():
            chrom,pos,ref,alt,maf,gene,func,vclass,aachange,metasvm,fathmm = [i for i in key]
            
            # Determine if probandGT quality was high enough
            proband,mom,dad = (value[0])
            pid,gt,ad,gq = proband.split(':')
            mgq = mom.split(':')[-1]
            dgq = dad.split(':')[-1]
            
            if any(i < 80 for i in [float(gq), float(mgq), float(dgq)]):
                continue
            
            #if float(gq) < 90:
            #    continue
            
            if func == "exonic":
                if vclass == "nonsynonymous_SNV":
                    if metasvm=='D' or fathmm=='D':
                        varclass = 'misD'
                    else:
                        varclass = 'mis'
                elif vclass == "synonymous_SNV":
                    varclass = 'syn'
                elif vclass in(['frameshift_deletion','frameshift_insertion']):
                    varclass = 'frameshift'
                elif vclass in(['stopgain']):
                    varclass = 'non'
                else:
                    print("hmmmm: ", vclass)
            elif 'splic' in func:
                varclass = 'splice'
            
            # If so, create list [Blinded ID, CHR, POS, REF, ALT, Gene, Variant Class, AA change, PGQ, MGQ, FGQ]
            dnv_list.append([pid, chrom, pos, ref, alt, gene, varclass, aachange, gq, mgq, dgq] )
```

```python
import pandas as pd
dnvs = pd.DataFrame(dnv_list, columns = ['Blinded ID', 'CHR', 'POS', 'REF', 'ALT', 'Gene',
                                         'Variant Class', 'AA change', 'PGQ', 'MGQ', 'FGQ'])
```

```python
## Compare to canonical natgen list

natgen = pd.read_csv('/pollard/home/mpittman/ctf-apms/data/variants/DNV_cases.csv')
natgen.shape
```

```python
natgen['POS'] = natgen['POS'].astype(int)
dnvs['POS'] = dnvs['POS'].astype(int)
combined = natgen.merge(dnvs, how="outer", on = ['Blinded ID','CHR','POS','REF','ALT'], indicator=True)
```

```python
natgen_only = combined.loc[combined['_merge']=='left_only']
natgen_only.shape
```

```python
new_only = combined.loc[combined['_merge']=='right_only']
new_only.shape
```

```python
both = combined.loc[combined['_merge']=='both']
both.shape
```

```python
print(both['MGQ'].min())
print(both['FGQ'].min())
print(both['PGQ'].min())
```

```python
# Find indiviuals who are new vs repeated
pids_ng = both['Blinded ID'].tolist() + natgen_only['Blinded ID'].tolist()
pids_new = both['Blinded ID'].tolist() + new_only['Blinded ID'].tolist()

pids_ng_only = [i for i in pids_ng if i not in pids_new]
pids_both = [i for i in pids_new if i in pids_ng]
pids_new_only = [i for i in pids_new if i not in pids_ng]

print(len(pids_ng_only))
print(len(pids_new_only))
print(len(pids_both)) 
```

```python
print(both.columns)
print(new_only.columns)
print(natgen_only.columns)
```

```python
col_list = ['Blinded ID', 'CHR', 'POS', 'REF', 'ALT', 'Gene','Variant Class', 'AA change']

# Save out
both['Gene'] = both['Gene_x']
both['Variant Class'] = both['Variant Class_x']
both['AA change'] = both['AA change_x']
both = both[col_list]

natgen_only['Gene'] = natgen_only['Gene_x']
natgen_only['Variant Class'] = natgen_only['Variant Class_x']
natgen_only['AA change'] = natgen_only['AA change_x']
natgen_only = natgen_only[col_list]

new_only['Gene'] = new_only['Gene_y']
new_only['Variant Class'] = new_only['Variant Class_y']
new_only['AA change'] = new_only['AA change_y']
new_only = new_only[col_list]
```

```python
final = pd.concat([both,natgen_only,new_only])
final.to_csv("/pollard/home/mpittman/ctf-apms/data/variants/all_DNV_cases.csv", index=False)
```
